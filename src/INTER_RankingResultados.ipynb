## Obtener Indicadores de Rendimiento para las Farmacias

Usando el año  mas reciente 

```python
farmDatos = farmDatos[farmDatos["Año Censal"] == 2018]
```

```python
farmDatos['ROE'] = (farmDatos['A231A Margen bruto de operación (Porcentaje)'] / 100) * farmDatos['M000A Total de ingresos por suministro de bienes y servicios (millones de pesos)']
farmDatos['ROA'] = (farmDatos['A131A Valor agregado censal bruto (millones de pesos)'] / farmDatos['Q000A Acervo total de activos fijos (millones de pesos)']) * 100
farmDatos['Rentabilidad sobre ingresos'] = (farmDatos['A231A Margen bruto de operación (Porcentaje)'] / 100) * farmDatos['M000A Total de ingresos por suministro de bienes y servicios (millones de pesos)'] / farmDatos['M000A Total de ingresos por suministro de bienes y servicios (millones de pesos)'] * 100
farmDatos['Gastos operativos'] = farmDatos['A700A Total de gastos (millones de pesos)'] / farmDatos['M000A Total de ingresos por suministro de bienes y servicios (millones de pesos)'] * 100
farmDatos['Gastos por farmacia'] = farmDatos['A700A Total de gastos (millones de pesos)'] / farmDatos['UE Unidades económicas']
farmDatos['Ingresos por farmacia'] = farmDatos['M000A Total de ingresos por suministro de bienes y servicios (millones de pesos)'] / farmDatos['UE Unidades económicas']
farmDatos['Remuneración media'] = farmDatos['J000A Total de remuneraciones (millones de pesos)'] / farmDatos['UE Unidades económicas']
farmDatos['Gasto en publicidad por farmacia'] = farmDatos['K810A Gastos por publicidad (millones de pesos)'] / farmDatos['UE Unidades económicas']
farmDatos['Inventario total'] = (farmDatos['P000A Total de inventario inicial (millones de pesos)'] + farmDatos['P000B Total de inventario final (millones de pesos)']) / 2
farmDatos['Valor agregado por farmacia'] = farmDatos['A131A Valor agregado censal bruto (millones de pesos)'] / farmDatos['UE Unidades económicas']
farmDatos['Variación de existencias'] = farmDatos['P000C Variación total de existencias (millones de pesos)'] / farmDatos['M000A Total de ingresos por suministro de bienes y servicios (millones de pesos)'] * 100
farmDatos['Depreciación por farmacia'] = farmDatos['Q000B Depreciación total de activos fijos (millones de pesos)'] / farmDatos['UE Unidades económicas']
farmDatos['Eficiencia en costos'] = (farmDatos['A121A Consumo intermedio (millones de pesos)'] / farmDatos['A131A Valor agregado censal bruto (millones de pesos)']) * 100
```

```python
#EFICIENCIA
# Define weights for each parameter
weights = {
    'ROE': 0.25,                       # Weight for Return on Equity
    'ROA': 0.20,                       # Weight for Return on Assets
    'Rentabilidad sobre ingresos': 0.15, # Weight for Profitability over Income
    'Gastos operativos': -0.10,        # Weight for Operating Expenses (negative, as lower is better)
    'Gastos por farmacia': -0.05,      # Weight for Expenses per Pharmacy (negative)
    'Ingresos por farmacia': 0.15,     # Weight for Income per Pharmacy
    'Remuneración media': 0.05,        # Weight for Average Salary
    'Gasto en publicidad por farmacia': -0.05, # Weight for Advertising Expense per Pharmacy (negative)
    'Inventario total': 0.05,          # Weight for Total Inventory
    'Valor agregado por farmacia': 0.05, # Weight for Value Added per Pharmacy
    'Variación de existencias': -0.05, # Weight for Change in Inventory (negative)
    'Depreciación por farmacia': -0.05, # Weight for Depreciation per Pharmacy (negative)
    'Eficiencia en costos': 0.10        # Weight for Cost Efficiency
}

# Calculate efficiency as a weighted sum
farmDatos['Eficiencia'] = (
    farmDatos['ROE'] * weights['ROE'] +
    farmDatos['ROA'] * weights['ROA'] +
    farmDatos['Rentabilidad sobre ingresos'] * weights['Rentabilidad sobre ingresos'] +
    farmDatos['Gastos operativos'] * weights['Gastos operativos'] +
    farmDatos['Gastos por farmacia'] * weights['Gastos por farmacia'] +
    farmDatos['Ingresos por farmacia'] * weights['Ingresos por farmacia'] +
    farmDatos['Remuneración media'] * weights['Remuneración media'] +
    farmDatos['Gasto en publicidad por farmacia'] * weights['Gasto en publicidad por farmacia'] +
    farmDatos['Inventario total'] * weights['Inventario total'] +
    farmDatos['Valor agregado por farmacia'] * weights['Valor agregado por farmacia'] +
    farmDatos['Variación de existencias'] * weights['Variación de existencias'] +
    farmDatos['Depreciación por farmacia'] * weights['Depreciación por farmacia'] +
    farmDatos['Eficiencia en costos'] * weights['Eficiencia en costos']
)
```

```python
municipios_ranking_sorted = municipios_ranking[['Entidad','Municipio','Distancias_Rank','Ranking_Ef', 'Difference_Rank','Ingresos_Rank','Difference','Ingresos por farmacia', 'Farms_x_pop']].sort_values(by=['Distancias_Rank', 'Ranking_Ef', 'Difference_Rank','Ingresos_Rank' ], ascending=[True, True, True, False])
municipios_ranking_sorted.head(60)
```

### Ranking cumulativo de los 4 indicadores

```python
municipios_ranking_sorted['Cumulative_Rank'] = municipios_ranking_sorted[['Ranking_Ef', 'Difference_Rank', 'Distancias_Rank', 'Ingresos_Rank']].sum(axis=1)
municipios_ranking_sorted.sort_values('Cumulative_Rank').head(10)
```

```python
mean_difference = municipios_ranking_sorted['Difference'].mean()

mean_difference
```

```python
municipios_ranking_sorted['Normalized_Difference'] = (municipios_ranking_sorted['Difference'] - municipios_ranking_sorted['Difference'].min()) / (municipios_ranking_sorted['Difference'].max() - municipios_ranking_sorted['Difference'].min())
```

```python
municipios_ranking_sorted.sort_values('Cumulative_Rank').head(10)
```

42 se usa para aplanar 

```python
municipios_ranking_sorted['Rec_Farms'] = municipios_ranking_sorted['Normalized_Difference'] * 42

municipios_ranking_sorted['Rec_Farms'] = municipios_ranking_sorted['Rec_Farms'].round(0)

municipios_ranking_sorted = municipios_ranking_sorted.sort_values('Cumulative_Rank')
municipios_ranking_sorted = municipios_ranking_sorted.sort_values('Cumulative_Rank')

municipios_ranking_sorted['Cumulative_Product'] = municipios_ranking_sorted['Rec_Farms'].cumsum()
filtered_rows = municipios_ranking_sorted[municipios_ranking_sorted['Cumulative_Product'] <= 200]
final = filtered_rows[filtered_rows['Rec_Farms'] > 0]
```

```python
correlation_df = filtered_farmDatos.groupby(['Entidad', 'Municipio'])[['Año Censal', 'P000C Variación total de existencias (millones de pesos)']].corr().unstack()['P000C Variación total de existencias (millones de pesos)']['Año Censal']

correlation_df = correlation_df.dropna()

negative_correlation_threshold = -0.7
strong_negative_correlations = correlation_df[correlation_df <= negative_correlation_threshold]

municipalities_negative_correlation = strong_negative_correlations.index.tolist()
```

```python
common_municipalities = set(municipalities_negative_correlation) & set(municipalities_positive_correlation)

print("\nMunicipalities with strong negative correlation in 'Variación total de existencias' and strong positive correlation in 'Ingresos':")
print(list(common_municipalities))
```
